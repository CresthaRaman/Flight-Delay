Without Normlaizing the data, we can see that it is very difficult to find any patterns in the data. We will need to normalize the data before we can start to find any patterns.Before we normalize the data, we will need to check for any missing values in the data. We can use the isnull() function to check for any missing values in the data. We can also use the sum() function to count the number of missing values in each column.

1. Check for missing values

```python
df.isnull().sum()   
```

```python
for column in df.columns:
    if df[column].dtype in ['int64', 'float64']:
            df[column].fillna(df[column].mean(), inplace=True)

```

```python
df['arr_flights'].isnull().sum()
```

```python
sns.boxplot(y=df['arr_flights'])
plt.title("Box Plot of arr_flights")
plt.show()
```

```python
sns.boxenplot(y=df['arr_flights'], color='lightblue', showfliers=True)
plt.title("Boxen Plot of arr_flights")
plt.show()
```

```python
df.describe(percentiles=[0.25, 0.5, 0.75, 0.95, 0.99])
```

```python
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr_value = q3 - q1
        lower_bound = q1 - 1.5 * iqr_value
        upper_bound = q3 + 1.5 * iqr_value
        df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
sns.boxenplot(y=df['arr_flights'], color='lightblue')
plt.title("Boxen Plot of arr_flights")
plt.show()
```
